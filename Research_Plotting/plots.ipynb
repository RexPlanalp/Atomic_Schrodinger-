```python
import os
import numpy as np
import matplotlib.pyplot as plt

path = "/data/becker/dopl4670/old_TDSE_Jobs/Project/Frequency_Sweep"

# Create a 1x2 grid of subplots
fig, axs = plt.subplots(1, 2, figsize=(20, 10))

count = 0
# Loop over all first-generation subdirectories
for item in os.listdir(path):
    item_path = os.path.join(path, item)
    # Check if it's a directory and not 'base'
    if os.path.isdir(item_path) and item != 'base':
        parts = item.split('_')
        rounded_number = round(float(parts[1]), 3)
        y = np.load(os.path.join(item_path, "TDSE_files", "A_slice.npy"))

        if count <= 3:
            axs[0].plot(y, label=f"{rounded_number} nm")  # Add label for legend
        else:
            axs[1].plot(y, label=f"{rounded_number} nm")  # Add label for legend
        count += 1

# Show legends for each subplot
axs[0].legend()
axs[1].legend()

plt.tight_layout()
plt.show()

       
        
```
